```python
def describle_pet(animal_type:str, pet_name:str, age:int)->None:
    '''Docstring'''
    months = int(age*12)
    print(f'I have a {animal_type}, its name is {pet_name} and it is {months} months old')
```

```python
describle_pet('Dog', "Xoai", 2)
```

```python
describle_pet('xoai','Dog', 2)
```

```python
describle_pet(animal_type='Dog',age=2, pet_name='Xoai')
```

```python
def get_formatted_fullname(first_name:str, last_name:str)->str:
    full_name = f'{first_name} {last_name}'
    return full_name.title()
```

```python
mfname = get_formatted_fullname('nguyen', 'tuan long')
print(mfname)
```

```python
def is_prime(n:int)->bool:
    '''Check prime'''
    if n<= 1:
        return False
    for d in range(2,n):
        if n%d == 0:
            return False
    return True 
```

```python
is_prime(17)
```

```python
def build_profile(first: str, last:str, **user_info)->dict:
    """Docstring"""
    user_info['first'] = first
    user_info['LAST'] = last
    return user_info
```

```python
build_profile('Long','Nguyen', location = 'Hanoi', job = 'lector')
```
